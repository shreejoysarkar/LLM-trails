# Tensors

### Creating Tensors

```python
## scalar - tensor with zero dimension
tensor_0d = torch.tensor(1)
tensor_0d
```

```python
tensor_0d.ndim
```

```python
## Vector - tensor with one dimension
tensor_1d = torch.tensor([1,2,4])
tensor_1d
```

```python
tensor_1d.ndim
```

```python
## Matrix - tensor with two dimensions
tensor_2d = torch.tensor([[0,1,4],[4,12,6]])
tensor_2d
```

```python
## Dimension and shape of a tensor
print(f'vector:\n,{tensor_1d} \t No of dimensions: {tensor_1d.ndim} \t Shape: {tensor_1d.shape}')
print(f'matrix:\n,{tensor_2d}\t No of dimensions: {tensor_2d.ndim} \t Shape: {tensor_2d.shape}')
```

```python
## Creating tensors from a numpy array
import numpy as np

arr = np.array([[52,5],[24,5]])
tensor8 = torch.from_numpy(arr)

tensor9 = torch.tensor(arr)
print(arr)
print(tensor8)
print(tensor9)
```

```python
arr *= 3
print(arr)
print(tensor8)
print(tensor9)
```

```python
###creating a tensor from another tensor

tensor10 = torch.ones_like(tensor8)
tensor10
```

# Basic Tensor Operation

```python
tensor12 = torch.ones(2,3)
tensor13 = torch.rand(2,3)
```

```python
print(tensor12,'\n', tensor13)
```

```python
# Element wise operation

# addtion
tensor14 = tensor12+tensor13
#   tensor.add(tensor12,tensor13)

# substration
tensor15 = tensor12-tensor13
#   tensor.sub(tensor12,tensor13)

# multiplication
tensor16 = tensor12*tensor13
#   tensor.mul(tensor12,tensor13)

# division
tensor17 = tensor12/tensor13
#   tensor.div(tensor12,tensor13)

print(tensor14,'\n', tensor15, '\n', tensor16, '\n', tensor17)
```
